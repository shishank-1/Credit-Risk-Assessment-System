```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')
df.head()
```

```python
df.columns
```

2Ô∏è‚É£ Loan Status Distribution

```python
sns.countplot(x='Loan_Status', data=df, palette='viridis')
plt.title('Loan Approval vs Rejection')
plt.show()
```

3Ô∏è‚É£ Credit History vs Loan Status-Ye clearly dikhaega ke Credit_History = 1 wale logon ka approval chance kitna zyada hai.

```python
sns.countplot(x='Credit_History', hue='Loan_Status', data=df, palette='Set2')
plt.title('Impact of Credit History on Loan Approval')
plt.show()
```

4Ô∏è‚É£ Income Distribution- üëâ Income ka pattern aur range samajhne me help karta hai.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ApplicantIncome'], bins=30, kde=True)
plt.title('Applicant Income Distribution')
plt.show()
```

5Ô∏è‚É£ Property Area Impact - Dekho urban, semiurban, rural me loan approval rate ka difference.

```python
sns.countplot(x='Property_Area', hue='Loan_Status', data=df, palette='cool')
plt.title('Loan Approval by Property Area')
plt.show()
```

6Ô∏è‚É£ Correlation Heatmap - Ye dikhaata hai kaunse numeric columns strongest correlation rakhte hain Loan_Status se.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Features')
plt.show()
```
